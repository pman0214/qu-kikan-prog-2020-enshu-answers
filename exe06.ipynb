## 演習問題

```python
# ブロックサイズを入力
block_size = int(input("モザイクブロックサイズは？: "))

# 画素値を入力
pixels = []
for i in range(block_size**2):
  pixels.append(int(input("画素{}の画素値: ".format(i))))

# 画素値を出力
for i in range(block_size):
  for j in range(block_size):
    print("{:5d}".format(pixels[i*block_size+j]), end="")
  print("")

# 代表画素値を計算（小数点以下は切り捨て）
rep_pixel = int(sum(pixels) / len(pixels))
print("代表画素値: {}".format(rep_pixel))
```

## 発展問題

```python
# データ
import numpy as np
x = np.linspace(0, 20*np.pi, 200)
data = np.sin(x) * np.exp(-x/len(x)*10)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, data)
plt.xlabel('x')
plt.ylabel('Diminish sin(x)')
plt.xlim([0,np.max(x)])
plt.ylim([-1,1])
# ファイルに保存したいときにはsavefigを使う
plt.savefig('exe06-2.png', dpi=600)
```
