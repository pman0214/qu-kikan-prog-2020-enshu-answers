## 演習問題

```python
# モザイクブロックのサイズ
block_size = 50

# 画像サイズ
img_size = [800, 500]
```

```python
top_lefts = []
for x in range(0, img_size[0], block_size):
    for y in range(0, img_size[1], block_size):
        top_lefts.append("({},{})".format(x, y))
```

```python
len(top_lefts)
```

```python
print(", ".join(top_lefts))
```

## 発展問題

```python
x = ['赤い', '青い', '黄色い', '白い']
y = ['ボール', '鉛筆', '花']
z = ['投げる', '捨てる', '買う']
```

```python
import itertools
patterns = list(itertools.product(x, y, z))
patterns
```

```python
len(patterns)
```

```python
print(",\n".join(list(map(lambda x: str(x), patterns))))
```
